### 5.2. For 

image.png

image.png

*Note:* sequence and iterator is the same. 
common iterator in python:
- string.
- list.
- tuple.
- set.
- dictionary.

```python
name = 'Jisoo'
for i in name:
  print(i)
```

```python
lst = ['Jisoo', 'Jennie', 'Lisa', 'Rose']
for name in lst:
  print(name)
```

```python
dct = {'member 1': 'Jisoo', 
       'member 2': 'Jennie',
       'member 3': 'Lisa', 
       'member 4': 'Rose'}
for i in dct:
  print(i, dct[i])
```

For usually go along with `range()`

image.png

```python
for i in range(5):
  print(i)
```

```python
for i in range(1, 5):
  print(i)
```

```python
for i in range(1, 5, 2):
  print(i)
```

```python
lst = ['Jisoo', 'Jennie', 'Lisa', 'Rose']
for idx in range(len(lst)):
  print(f"idx: {idx}")
  print(f"lst[{idx}]: {lst[idx]}")
```

### 5.4. Break, continue, pass in Loop satement

#### 5.4.1. Break

image.png

image.png

```python
a = 5
while True:
  print(f'a = {a}')
  a -= 1
  if a == 0:
    break
print('end loop')
```

```python
a = 5
while True:
  print(f'a = {a}')
  if a == 0:
    break
  a -= 1
print('end loop')
```

#### 5.4.2. Continue

image.png

image.png

```python
a = 6
while a > 0:
  a -= 1
  if a % 2 == 0:
    continue
  print(f'a = {a}')
print('end loop')
```

#### 5.4.3. Pass

image.png

image.png

```python
a = 6
while a > 0:
  a -= 1
  if a % 2 == 0:
    print('this a is even')
  print(f'a = {a}')
print('end loop')
```

```python
a = 6
while a > 0:
  a -= 1
  if a % 2 == 0:
    pass
  print(f'a = {a}')
print('end loop')
```

### 5.5. Use case of For and while loop

- For loop use when you know number of iteration.

*Note:* In Python you can use For to go through the iterator.

- While loop use when you don't really know number of iteration.
Usually go along with `break`.
