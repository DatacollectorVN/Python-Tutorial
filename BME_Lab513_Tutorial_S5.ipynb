# 1. TXT file 

## 1.1 Read

````python
txt_file = 'data/sample.txt'
f = open(txt_file, 'r')
result = f.read()
print(result)
````

*Note:* `r` is mode read.

Read more [here](https://www.programiz.com/python-programming/methods/built-in/open)

````python
# must close file 
f.close()
````

Use `with open` to automatical closing file.

````python
with open(txt_file, 'r') as f:
  result = f.read()
  print(result)
````

## 1.2 Write

````python
lines = ['Readme\n', 'How to write text files in Python\n']
with open(txt_file, 'a+') as f:
  f.write('\n')
  f.write('first line')
  f.writelines(lines)
````

````python
with open(txt_file, 'r') as f:
  result = f.read()
  print(result)
````

## 2.1 Read

````python
csv_file = 'data/sample.csv'
with open(csv_file, 'r') as f:
  rows = csv.reader(f)
  for row in rows:
    print(row)
````

## 2.2 Write

````python
with open(csv_file, 'a+') as f:
  writer = csv.writer(f, delimiter=',')
  writer.writerow(['jennie', 26])
  writer.writerow(['lisa', 25])
  writer.writerow(['rose', 25])
````
