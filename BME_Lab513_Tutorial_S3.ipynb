```python
from datetime import datetime
```

```python
start = datetime.now()
for i in range(100):
  pass
end = datetime.now()
duration = end - start
print(f'duration: {duration}')
```

```python
start = datetime.now()
for i in range(100):
  for j in range(100):
    pass
end = datetime.now()
duration = end - start
print(f'duration: {duration}')
```

```python
start = datetime.now()
for i in range(100):
  for j in range(100):
    for k in range(100):
      pass
end = datetime.now()
duration = end - start
print(f'duration: {duration}')
```

Comment: you can see if we use 3 nested loop in python --> it growth the computing cost --> we need to seperate it into function

```python
def loop_func():
  for x in range(100):
    pass
```

```python
start = datetime.now()
for i in range(100):
  for j in range(100):
    loop_func()
    
end = datetime.now()
duration = end - start
print(f'duration: {duration}')
```
