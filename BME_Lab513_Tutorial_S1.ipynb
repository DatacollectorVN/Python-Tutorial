## 2. print() function

image.png

Concatenate string by `,` or `+`.

````python
print('Nathan', 'Ngo')
print('Nathan' + ' ' + 'Ngo')
````



```
print('Nathan', 'Ngo')
```
Why this output have a space between 2 words ?

*Output:* `Nathan Ngo`

Because the `sep` parameter in print() function with default value is `' '`.

If we change it default value as:


````python
print('Nathan', 'Ngo', sep = ' ')
print('Nathan', 'Ngo', sep = '')
print('Nathan', 'Ngo', sep = ',')
print('Nathan', 'Ngo', sep = '+')
````

If you don't want to enter the new line in two seperate print() functions --> use `end` operator.

The `end` operator have default value is `\n` is create the newline, so if we change it default value as:

````python
print('Nathan', 'Ngo', end = '\t')
print('Kim', 'Jisoo')
````

````python
print('Nathan', 'Ngo', end = ' ')
print('Kim', 'Jisoo')
````

````python
print('Nathan', 'Ngo', end = ' <3 ')
print('Kim', 'Jisoo')
````

## 3. print() function with variable

What is variable in Python ?. In this lecture we just showed you a simple variable type for illustrating, the detail we will learn a next lecture.

Some common variable type:
- integer (int)
- floating-point (float)
- string (str)

````python
# integer 
a = 5
print(type(a))
````

````python
# floating-point 
a = 5.5
print(type(a))
````

````python
# string 
a = "Nathan Ngo"
# or
b = 'Nathan Ngo'

print(type(a))
print(type(b))
````

print() with variables:
- Normal way
- Use `%`
- Use `{}.format()`
- Use `f"{}"`


````python
a = 5
b = 5.5678
# normal way
print('a =', a)
print('b =', b)

# Use %
print('a = %d' % (a))
print('b = %.2f' % (b))
print('b = %.3f' % (b))

# Use {}.format
print('a = {}'.format(a))
print('b = {:.3f}'.format(b))

# Use f{""}
print(f'a = {a}')
print(f'b = {b:.2f}')
````

````python
a = 5
b = 6.5
print(f'{a} + {b:.2f} = {a+b:.2f}')
````

````python
a = 'Nathan'
b = 'Ngo'
# use double quote inside sing quote (if use the same --> get error)
print(f'{a + " " + b}') 
````

Let see the example below:

````python
phy = input('Enter your physic score: ')
math = input('Enter your math score: ')
print(f"Your Total score is {phy + math}")
````

Why the result {phy + math} = `80.590`. 
Because the input() function always return value with string type even you type string or number.

````python
phy = input('Enter your physic score: ')
math = input('Enter your math score: ')
print(type(phy))
print(type(math))
````

To get the complier infer the type of input value --> use `eval(input())`

````python
phy = eval(input('Enter your physic score: '))
math = eval(input('Enter your math score: '))
print(f"Your Total score is {phy + math}")
print(type(phy))
print(type(math))
````
